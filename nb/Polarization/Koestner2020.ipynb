## Plot

```python
plt.clf()
ax = plt.gca()

nsample = p22_samples.shape[1]
for kk in range(nsample):
    ax.plot(psis, p22_samples[:,kk], color='k', alpha=0.2 + 0.8*kk/(nsample-1))

# Median
ax.plot(psis, p22_median, color='r', label='median')
ax.set_xlabel(r'$\psi$ [deg]')
ax.set_ylabel(r'$p_{22}$')
#ax.set_xlim(10, 155.)
#ax.set_yscale('log')
ax.legend()
plt.show()
```
