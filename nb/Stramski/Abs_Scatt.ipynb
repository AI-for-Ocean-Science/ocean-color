# Absorption and Scattering

```python
# imports
from importlib import reload
import numpy as np

from matplotlib import pyplot as plt

from oceancolor import cross
```

# Detritus, Minerals and Bubbles

```python
wave = np.linspace(350., 800., 1000)
```

## Absorption and Scattering

```python
reload(cross)
```

```python
d_abs = cross.detritus_abs(wave)
m_abs = cross.minerals_abs(wave)
```

```python
d_scatt = cross.detritus_scatt(wave)
m_scatt = cross.mineral_scatt(wave)
```

### Plot

```python
plt.clf()
ax = plt.gca()

# Absorption
ax.plot(wave, d_abs/1e12, color='r', label=r'Detritus $\sigma_a$')
ax.plot(wave, m_abs/1e12, color='b', label=r'Mineral $\sigma_a$')

# Scattering
ax.plot(wave, d_scatt/1e12, color='r', ls='--', label=r'Detritus $\sigma_b$')
ax.plot(wave, m_scatt/1e12, color='b', ls='--', label=r'Mineral $\sigma_b$')


ax.set_ylabel(r'Cross-section (m$^2$ per particle)')
ax.set_yscale('log')

ax.legend()

plt.show()
```

## Backscattering

```python
d_bscatt = cross.detritus_backscatt(wave)
m_bscatt = cross.mineral_backscatt(wave)
```

```python
plt.clf()
ax = plt.gca()

# Backscattering
ax.plot(wave, d_bscatt/1e12, color='r',  label=r'Detritus $\sigma_{bs}$')
ax.plot(wave, m_bscatt/1e12, color='b',  label=r'Mineral $\sigma_{bs}$')


ax.set_ylabel(r'Cross-section (m$^2$ per particle)')
ax.set_yscale('log')

ax.legend()

plt.show()
```
